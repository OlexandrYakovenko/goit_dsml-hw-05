Створіть нову таблицю python_data, в якій будуть тільки рядки зі спеціалістами, які вказали мову програмування Python

Використовуючи метод groupby, виконайте групування за стовпчиком "Посада"

```python
# Виконання групування за стовпчиком 'Посада'
grouped_df = python_data.groupby('Посада')

# # Виведення перших рядків кожної групи
# for position, group in grouped_df:
#     print("Група:", position)
#     print(group)
```

Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата.в.місяць"

```python
# Виконання групування за стовпчиком 'Посада' і агрегація даних
aggregated_df = python_data.groupby('Посада').agg({'Зарплата на місяць': ['min', 'max']})

# Перейменування стовпців у новому DataFrame
aggregated_df.columns = ['Мінімальна_зарплата', 'Максимальна_зарплата']

# Виведення нового DataFrame
print(aggregated_df)
```

Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. Використовуйте її для методу apply та створіть новий стовпчик "avg"

```python
# Функція для обчислення середнього значення заробітної плати на місяць
def fill_avg_salary(row):
    return (row['Мінімальна_зарплата'] + row['Максимальна_зарплата']) / 2

# Створення нового стовпця 'avg' за допомогою методу apply
aggregated_df['avg'] = aggregated_df.apply(fill_avg_salary, axis=1)#, axis=1

# Виведення DataFrame з новим стовпцем 'avg'
print(aggregated_df.head())
```

Створіть описову статистику за допомогою методу describe для нового стовпчика.

```python
# Описова статистика для нового стовпчика 'Row_Sum'
describe_stats = aggregated_df['avg'].describe()

# Виведення результату
print(describe_stats)
```

Збережіть отриману таблицю в CSV файл

```python
# Збереження описової статистики у CSV файл
describe_stats.to_csv('describe_stats.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Стовпчикова діаграма (Bar chart)
plt.figure(figsize=(8, 6))
sns.barplot(x=aggregated_df.index, y=aggregated_df['Мінімальна_зарплата'], palette='viridis')
plt.title('Розподіл даних за колонкою 1')
plt.xlabel('Категорії')
plt.ylabel('Значення')
plt.xticks(rotation=45)
plt.show()

# 2. Лінійний графік (Line plot)
plt.figure(figsize=(8, 6))
sns.lineplot(data=aggregated_df, x='Посада', y='Мінімальна_зарплата', marker='o', color='skyblue')
plt.title('Зміна значення в часі')
plt.xlabel('Час')
plt.ylabel('Значення')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# 3. Кругова діаграма (Pie chart)
plt.figure(figsize=(8, 6))
aggregated_df.index.value_counts().plot(kind='pie', autopct='%1.1f%%', colors=sns.color_palette('Set3'))
plt.title('Відносний розподіл категорій')
plt.ylabel('')
plt.show()

# 4. Ящик з вусами (Box plot)
plt.figure(figsize=(8, 6))
sns.boxplot(data=aggregated_df, x='Посада', y='Мінімальна_зарплата', palette='pastel')
plt.title('Розподіл даних за категоріями')
plt.xlabel('Категорії')
plt.ylabel('Значення')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# 5. Точкова діаграма (Scatter plot)
plt.figure(figsize=(8, 6))
sns.scatterplot(data=aggregated_df, x='Посада', y='Мінімальна_зарплата', hue='Максимальна_зарплата', palette='Set1', s=100)
plt.title('Залежність між двома змінними')
plt.xlabel('Змінна 1')
plt.ylabel('Змінна 2')
plt.legend(title='Категорія')
plt.grid(True)
plt.show()
```
